```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv')
df.head()
```

```python
df=df.set_index(df.columns[0])
df.head()
```

```python
df.shape
```

```python
dic = {
    'Unknown'        : 'N',
    'Less than $40K' : 'a',
    '$40K - $60K'    : 'b',
    '$60K - $80K'    : 'c',
    '$80K - $120K'   : 'd',
    '$120K +'        : 'e'
}
df['newIncome']=df.Income_Category.map(lambda x: dic[x])
df['newIncome'].head()
```

```python
df.reset_index(drop=True, inplace=True)
df['Customer_Age'].head()
```

```python
df['AgeState']=df.Customer_Age.map(lambda x: (x//10) *10)
result = df['AgeState'].value_counts().sort_index()
result.head(10)
```

```python
df['newEduLevel']=df.Education_Level.map(lambda x : 1 if 'Graduate' in x else 0)
result = df['newEduLevel'].value_counts()
result.head()
```

```python
df['newLimit']=df.Credit_Limit.map(lambda x : 1 if x>=4500 else 0)
result = df['newLimit'].value_counts()
result.head()
```

```python
def check(x):
    if x.Marital_Status == 'Married' and x.Card_Category=='Platinum':
        return 1
    else:
        return 0
df['newState']=df.apply(check,axis=1)
result = df['newState'].value_counts()
result.head()
```

```python
def gender_trans(x):
    if x.Gender == 'M':
        return 'male'
    if x.Gender == 'F':
        return 'female'
    else:
        return x.Gender

df['Gender'] = df.apply(gender_trans, axis=1)
result = df['Gender'].value_counts()
result.head()
```
