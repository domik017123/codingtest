```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score

#데이터 로드
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/X_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/y_train.csv")
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/X_test.csv")


drop_col = ['Surname','Geography','Gender']
x_train_drop = x_train.drop(columns = drop_col)
x_test_drop = x_test.drop(columns = drop_col)

x_train_drop.fillna(0)
x_test_drop.fillna(0)
y = y_train['Exited']

X_train, X_val, Y_train, Y_val = train_test_split(x_train_drop, y, test_size=0.33, random_state=42)
rf = RandomForestClassifier(random_state =23)
rf.fit(X_train,Y_train)

#model_score
pre_X_label = rf.predict(X_train)
pre_X_proba = rf.predict_proba(X_train)[:,1]

pre_X_val_label = rf.predict(X_val)
pre_X_val_proba = rf.predict_proba(X_val)[:,1]

print('train auc :', roc_auc_score(Y_train,pre_X_proba))
print('validation auc :', roc_auc_score(Y_val,pre_X_val_proba))


# test데이터 마찬가지 위와 같은 방식
pre_X_test_label = rf.predict(x_test_drop)
pre_X_test_proba = rf.predict_proba(x_test_drop)[:,1]
```
